```python
path = '/content/gdrive/MyDrive/Dataset.csv'
```

```python
import pandas as pd
```

```python
df = pd.read_csv(path)
df.head()
```

```python
correlation_matrix = df.corr()


correlation_with_energy = correlation_matrix['Energy Consumption (kWh)']

sorted_correlation = correlation_with_energy.sort_values(ascending=False)


print(sorted_correlation)
```

```python
column_data=df["Building Type"]
column_data
```

```python
df = pd.get_dummies(df, columns=["Building Type"], prefix=['Building Type'])
```

```python
df = pd.get_dummies(df, columns=['HVAC System'], prefix=['HVAC System'])
```

```python
column_data = df["HVAC System_ Split AC"]
```

```python
column_data
```

```python
df.isnull()
```

```python
df.columns
```

```python
X = df.drop(columns=['Energy Consumption (kWh)'])
y = df['Energy Consumption (kWh)']
```

```python
X.columns
```

```python
print(X.head(5))
```

```python
print(y.head(5))
```

```python
from sklearn import linear_model
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

y_pred = regr.predict(X)

mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y, y_pred)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R^2) Score: {r2:.2f}")
```
